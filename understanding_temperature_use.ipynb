# Temperature = control randomness of output

#### Low → accurate
#### High → creative

#### Temperature controls how “random” or “creative” the AI is while choosing words.

#### Low → safe, predictable
#### High → creative, surprising

#### Think: picking words from a list

#### Low temp = always pick top choice
#### High temp = sometimes pick unusual choices

```python
from openai import OpenAI
```

```python
f=open("keys/.openai_api_key.txt")
OPENAI_API_KEY=f.read()
client=OpenAI(api_key=OPENAI_API_KEY)
```

```python
def get_completion(prompt,model="gpt-4o-mini"):
    message=[{"role":"user","content":prompt}]
    response=client.chat.completions.create(
        model=model,
        messages=message,
        temperature=0
    )
    return response.choices[0].message.content
```

```python
response = get_completion(prompt="write a poem on india in 5 lines")
print(response)
```
